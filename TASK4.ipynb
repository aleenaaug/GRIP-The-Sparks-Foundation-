```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
data = pd.read_csv("/Users/HP/Downloads/globalterrorismdb_0718dist.csv",encoding='latin1')
data.head()
```

```python
data.columns.values
```

```python
data.rename(columns={'iyear':'Year','imonth':'Month','iday':"day",'gname':'Group','country_txt':'Country','region_txt':'Region','provstate':'State','city':'City','latitude':'latitude',
    'longitude':'longitude','summary':'summary','attacktype1_txt':'Attacktype','targtype1_txt':'Targettype','weaptype1_txt':'Weapon','nkill':'kill',
     'nwound':'Wound'},inplace=True)
```

```python
data = data[['Year','Month','day','Country','State','Region','City','latitude','longitude',"Attacktype",'kill',
               'Wound','target1','summary','Group','Targettype','Weapon','motive']]
data.head()
```

```python
data.shape
```

```python
data.isnull().sum()
```

```python
data['Wound'] = data['Wound'].fillna(0)
data['kill'] = data['kill'].fillna(0)
```

```python
data['Casualities'] = data['kill'] + data['Wound']
```

```python
data.info()
```

```python
data.describe()
```

```python
year = data['Year'].unique()
years_count = data['Year'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (28,15))
sns.barplot(x = year,
           y = years_count,
           palette = "tab10")
plt.xticks(rotation = 50)
plt.xlabel('Year',fontsize=20)
plt.ylabel('Number of Attacks',fontsize=25)
plt.title('No:of Attack in each year',fontsize=30)
plt.show()
```

```python
pd.crosstab(data.Year, data.Region).plot(kind='area',stacked=False,figsize=(30,20))
plt.title('Terrorist Activities On The Basis Of Region(Year-Wise) ',fontsize=25)
plt.ylabel('Number of Attacks',fontsize=25)
plt.xlabel("Year",fontsize=25)
plt.show()
```

```python
attack = data.Country.value_counts()[:10]
attack
```

```python
data.Group.value_counts()[1:10]
```

```python
plt.subplots(figsize=(30,25))
sns.barplot(data['Country'].value_counts()[:10].index,data['Country'].value_counts()[:10].values,palette='YlOrRd')
plt.title('Mostly Affected Countries ')
plt.xlabel('Countries')
plt.ylabel('Number')
plt.xticks(rotation = 50)
plt.show()
```

```python
df = data[['Year','kill']].groupby(['Year']).sum()
fig, ax4 = plt.subplots(figsize=(30,20))
df.plot(kind='bar',alpha=0.7,ax=ax4)
plt.xticks(rotation = 50)
plt.title("People Died Due To Attack",fontsize=25)
plt.ylabel("Number of People got killed",fontsize=20)
plt.xlabel('Year',fontsize=25)
top_side = ax4.spines["top"]
top_side.set_visible(False)
right_side = ax4.spines["right"]
right_side.set_visible(False)
```

```python
data['City'].value_counts().to_frame().sort_values('City',axis=0,ascending=False).head(10).plot(kind='bar',figsize=(30,25),color='orange')
plt.xticks(rotation = 50)
plt.xlabel("City",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Top 10 most effected city",fontsize=20)
plt.show()
```

```python
data['Attacktype'].value_counts().plot(kind='bar',figsize=(30,25),color='brown')
plt.xticks(rotation = 50)
plt.xlabel("Attacktype",fontsize=15)
plt.ylabel("Number of attack",fontsize=15)
plt.title("Name of attacktype",fontsize=20)
plt.show()
```

```python
data[['Attacktype','kill']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(30,25),color=['cyan'])
plt.xticks(rotation=50)
plt.title("Number of killed ",fontsize=25)
plt.ylabel('Number of people',fontsize=20)
plt.xlabel('Attack type',fontsize=20)
plt.show()
```

```python
data[['Attacktype','Wound']].groupby(["Attacktype"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['cyan'])
plt.xticks(rotation=50)
plt.title("Number of wounded  ",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

```python
plt.subplots(figsize=(30,25))
sns.countplot(data["Targettype"],order=data['Targettype'].value_counts().index,palette="gist_heat",edgecolor=sns.color_palette("rocket"));
plt.xticks(rotation=90)
plt.xlabel("Attacktype",fontsize=25)
plt.ylabel("count",fontsize=25)
plt.title("Attack per year",fontsize=25)
plt.show()
```

```python
data['Group'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='blue',figsize=(30,20))
plt.title("Top 10 terrorist group attack",fontsize=25)
plt.xlabel("terrorist group name",fontsize=25)
plt.ylabel("Attack number",fontsize=25)
plt.show()
```

```python
data[['Group','kill']].groupby(['Group'],axis=0).sum().drop('Unknown').sort_values('kill',ascending=False).head(10).plot(kind='bar',color='red',figsize=(20,10))
plt.title("Top 10 terrorist group attack",fontsize=30)
plt.xlabel("terrorist group name",fontsize=25)
plt.ylabel("No of killed people",fontsize=25)
plt.show()
```
